# Data Analysis and Visualizations

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

This shows that about 62% of the bookings is not cancelled, while 37% of the booking are cancelled- which is really high as compared to the 62% not cancelled. So let's visualize this result using a bar chart

```python
plt.figure(figsize = (5,4))
plt.title('reservation_status_count')
plt.bar(['Not canceled', 'Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor(1,1))
plt.title('Reservation Status in Different Hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
df['is_canceled'] = df['is_canceled'].astype(str)
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

The above analysis quickly shows that for the Resort Hotel, while 72% of the bookings are not canceled, about 28% are actually being canceled.
However, for the City Hotel, 58% of the bookings are not canceled, while a whooping 42% of the total bookings are being canceled.

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rates in City and Resort Hotels', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df)
plt.xlabel = ('month')
plt.ylabel = ('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

From the bar graph, we can deduce that that highest cancellation rate occurred in January, and the lowest cancellation was done in August.
The reservation was about the same in the months of January and December.
It could also be that the reservations in August was the highest, with the lowest cancellation rate because the price was low. So let's confirm by plotting the average daily rate for each month.


```python
plt.figure(figsize = (15,8))
plt.title('ADR per Month', fontsize = 30)
sns.barplot(x = 'month', y = 'adr', data = df[df['is_canceled'].astype(float) == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data = df[df['is_canceled'].astype(float) == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 Countries With Reservations Cancelled')
plt.pie(top_10_countries, autopct = '%.2f', labels = top_10_countries.index)
plt.show()
```

From the Pie chart, we can see that Portugal has the highest cancellation rate of 70%. So recommendations for hotels in Portugal would be that they should increase/maintain the facilities in the hotel, reduce the price, do promotions and give discounts; also to run marketing campaigns and do advertisements.
Next, let's find out the channels where the clients are coming from to make the reservations.

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize = True)
```

From the above analysis, we can see that 47% of the clients are coming from the Online Travel Agency; While 20% of the clients are coming from the Offline Travel Agency.

```python
cancelled_data['market_segment'].value_counts(normalize = True)
```

From the above, we can see that most of the cancellations were done online, about 47%.

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'].astype(float) == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```
